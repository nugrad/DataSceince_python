```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
df=pd.read_csv('zoo.csv')
df
```

```python
# user func for predator
def origin(num):
    if num==1:
        return 'shikari'
    else:
        return 'non shikari'
df['predator']=df['predator'].apply(origin)
```
