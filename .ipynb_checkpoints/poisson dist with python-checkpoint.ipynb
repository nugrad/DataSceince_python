```python
from scipy.stats import poisson
from matplotlib import pyplot as plt
x=[0,1,2,3,4,5]
lmbda=2
# prob values
poisson_pd=poisson.pmf(x,lmbda)
# plot prob dist
fig,ax=plt.subplots(1,1)
ax.plot(x,poisson_pd,'bo',label='poisson pmf')
plt.ylabel('probabilities')
plt.xlabel('X= number of restaurants')
plt.title('Poisson Distribution : probability vs restaurants')
ax.vlines(x,0,poisson_pd,colors='b',alpha=0.5)
```
